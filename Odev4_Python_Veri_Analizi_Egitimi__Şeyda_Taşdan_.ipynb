```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import plotly.graph_objs as xaxis
from plotly.subplots import make_subplots
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
B30 = yf.download('XU030.IS', start = '2022-01-01' , end = '2022-12-31')
EUR_TRY = yf.download('EURTRY=X', start='2021-01-01', end='2021-12-31')
```

```python
B30 = yf.download('TRY=X', start = '2021-01-01' , end = '2021-12-31')
B30 = B30[B30.index>'2021-01-01']
B30 = B30[['Adj Close']]
B30.columns = ['Kapanış(TRY)']
B30['Normalize'] = B30['Kapanış(TRY)']/B30.iloc[0,0]

EUR_TRY = yf.download('TRY=X', start = '2021-01-01' , end = '2021-12-31')
EUR_TRY = EUR_TRY[EUR_TRY.index>'2021-01-01']
EUR_TRY = EUR_TRY[['Adj Close']]
EUR_TRY.columns = ['EUR/TRY']
EUR_TRY['Normalize EUR/TRY'] = EUR_TRY['EUR/TRY']/EUR_TRY.iloc[0,0]
```

```python
indirme_listesi=['XU100.IS' , 'TRY=X']
Veri = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start = '2021-01-01' , end = '2022-12-31')
  data = data['Adj Close'].rename(i)
  Veri = pd.concat([Veri, data], axis = 1)

Veri
```

```python
Veri.columns=['Kapanış(TRY)', 'EUR/TRY']
Veri = Veri[Veri['Kapanış(TRY)'] > 0 ]
Veri
```

### Veriyi özetleme

1.  3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/TRY kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir pivot tablo üretiniz.

```python
Veri['ay_yıl'] = Veri.index
Veri['ay_yıl'] = Veri['ay_yıl'].apply(lambda x: x.strftime('%Y-%m'))
Veri
```

```python
Veri1=Veri.pivot_table(values=['Kapanış(TRY)', 'EUR/TRY'], index='ay_yıl', aggfunc='last')
Veri1
```

3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/USD kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir tabloyu bir de resample yöntemi ile üretiniz.

```python
Veri1['degisim'] = Veri1['Kapanış(TRY)'].diff(1)
Veri1['d_yon'] = Veri1['degisim'].map(lambda x: 'düşen gün' if x<0 else 'yükselen gün')
Veri1['degisim%'] = (Veri1['degisim']/Veri['Kapanış(TRY)'].shift())
Veri1
```

```python
Veri.index = pd.to_datetime(Veri.index) 
```

resample fonksiyonunun yalnızca bazı tipteki dizinlere sahip DataFrame'lerle kullanılabileceği için Veri veri çerçevesinin indeksini bir tarih dizinine dönüştürmemiz gerekiyor. 
Bu kod bu işe yarıyor.

```python
Veri1 = Veri[['Kapanış(TRY)']].resample('BM').last()
Veri1['degisim'] = Veri1['Kapanış(TRY)'].diff(1)
Veri1['d_yon'] = Veri1['degisim'].map(lambda x: 'düşen gün' if x<0 else 'yükselen gün')
Veri1['degisim%'] = (Veri1['degisim']/Veri1['Kapanış(TRY)'].shift())
Veri1
```

4. Resample yöntemi ile ürettiğiniz tabloyu grafik olarak inceleyiniz. 2. adımda elde ettiğiniz grafik ile karşılaştırarak yorumlayınız.

```python
px.bar(Veri1, y='degisim',
       color='d_yon', color_discrete_map={'düşen gün' : 'red' , 'yükselen gün' : 'green'},
       text_auto='.0%')
```
