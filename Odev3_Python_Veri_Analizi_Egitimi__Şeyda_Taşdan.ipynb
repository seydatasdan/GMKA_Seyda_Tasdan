```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import plotly.graph_objs as xaxis
from plotly.subplots import make_subplots
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### 2021 EUR/TRY kur tablosunun incelenmesi

1. <i> Yahoo Finance</i> sitesinden 2021 yılına ait EUR/TRY kur değerlerini indiriniz

```python
EUR_TRY = yf.download('EURTRY=X', start='2021-01-01', end='2021-12-31')
EUR_TRY
```

2. Veriyi özet metodları ile inceleyiniz

```python
EUR_TRY.head()
```

```python
EUR_TRY.describe()
```

3. Veri temizleme: <br>
	• Tabloda sadece "Adj Close" (düzeltilmiş kapanış) sütununu bırakınız.<br>
	• "Adj Close" sütununu " Kapanış (TRY)" olarak yeniden adlandırınız.

```python
EUR_TRY = EUR_TRY[['Adj Close']]
EUR_TRY.columns=["Kapanış(TRY)"]
EUR_TRY
```

4. Tabloya yeni sütun ekleme:<br>
	• Tabloya yeni bir sütun ekleyerek Kapanış (TRY) değerlerini 1 Ocak 2021 tarihindeki kapanış değerine göre normalize ediniz.<br>
	• BIST30 endeksinin 2021 yılı boyunca değişimini yorumlayınız.

```python
norm = EUR_TRY.iloc[0,0]   # iloc belli bir konumdaki verileri seçmek için kullanılır
EUR_TRY['Normalize'] = EUR_TRY['Kapanış(TRY)'] / norm
EUR_TRY
```

BIST30 endeksi 2021 yılı boyunca artış göstermiştir.

5. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloyu px grafiği olarak görselleştiriniz.<br>
	• Tabloyu go. grafiği olarak görselleştiriniz.

```python
fig = px.line(EUR_TRY, y='Normalize', title='BIST30 Endeksi 2021 Yılı',
              labels = {'Date': 'Tarih', 'Normalize':'BIST 30 Endeksi'})
fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=EUR_TRY.index, y=EUR_TRY['Normalize'], 
                         mode='markers', name='BIST30 Endeksi'))

fig.update_layout(title='BIST30 Endeksi 2021 Yılı',
                  xaxis_title='Tarih',
                  yaxis_title='BIST30 Endeksi')

fig.show()
```

### 2021 ve 2022 değerlerini birlikte inceleme

1. <i> Yahoo Finance</i> sitesinden 2022 BIST30 ve EUR/TRY kur değerlerini indiriniz

```python
B30 = yf.download('XU030.IS', start = '2022-01-01' , end = '2022-12-31')
EUR_TRY = yf.download('EURTRY=X', start='2021-01-01', end='2021-12-31')
```

2. Indirdiğiniz verileri 2021 verilerini temizleyip düzenlediğiniz üzere hazırlayınız.

```python
B30 = yf.download('TRY=X', start = '2022-01-01' , end = '2022-12-31')
B30 = B30[B30.index>'2022-01-01']
B30 = B30[['Adj Close']]
B30.columns = ['Kapanış(TRY)']
B30['Normalize'] = B30['Kapanış(TRY)']/B30.iloc[0,0]
B30
```

```python
EUR_TRY = yf.download('TRY=X', start = '2022-01-01' , end = '2022-12-31')
EUR_TRY = EUR_TRY[EUR_TRY.index>'2022-01-01']
EUR_TRY = EUR_TRY[['Adj Close']]
EUR_TRY.columns = ['EUR/TRY']
EUR_TRY['Normalize EUR/TRY'] = EUR_TRY['EUR/TRY']/EUR_TRY.iloc[0,0]
EUR_TRY
```

3. 2021 ve 2022 veri tablolarını tek bir tablo haline getiriniz.

```python
B30 = B30.merge(EUR_TRY, left_index=True, right_index=True)
B30
```

4. Veriyi grafik olarak inceleyiniz:<br>
	• Tabloları px grafiği olarak görselleştiriniz.<br>
	• Tabloları go. grafiği olarak görselleştiriniz.<br>
	• Tabloları subplots grafiği olarak görselleştiriniz.

```python
fig = px.line(B30, y=['Normalize', 'Normalize EUR/TRY'])
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=EUR_TRY.index, y= EUR_TRY['Normalize EUR/TRY'] , name='Normalize EUR/TRY'))
fig.add_trace(go.Scatter(x=B30.index, y= B30['Normalize'], name = 'Normalize BIST 30'))

fig.update_layout(title = 'Normalize BIST 30 ve Normalize EUR/TRY Kur',
                  xaxis_title = 'Tarih', xaxis_dtick ='M1',
                  yaxis_title = 'Kapanış (TRY)')
```

```python
fig2 = make_subplots(rows=1, cols=2)

fig2.add_trace(
    go.Scatter(x=B30.index, y=B30['Normalize'], name="BIST 30"),
    row=1, col=1
)

fig2.add_trace(
    go.Scatter(x=EUR_TRY.index, y=EUR_TRY['Normalize EUR/TRY'], name="EUR/TRY"),
    row=1, col=2
)

fig2.update_layout(title="BIST 30 ve EUR/TRY Kur",
                  xaxis1_title="Tarih",
                  yaxis1_title="BIST 30 Kapanış (TRY)",
                  xaxis2_title="Tarih",
                  yaxis2_title="EUR/TRY Kapanış (TRY)")

fig2.show()

```
