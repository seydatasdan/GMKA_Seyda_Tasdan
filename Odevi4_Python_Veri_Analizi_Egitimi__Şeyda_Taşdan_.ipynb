```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import plotly.graph_objs as xaxis
from plotly.subplots import make_subplots
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Kur EUR 2021:
Kur = yf.download("EURTRY=X", start = "2021-01-01", end = "2023-01-01")
Kur = Kur[["Adj Close"]]
Kur.columns = ["EUR"]

# BIST30 2021:
B30 = yf.download("XU030.IS", start = "2021-01-01", end = "2023-01-01")
B30 = B30[["Adj Close"]]
B30.columns = ["BIST30"]
```

```python
B30_EUR = B30.merge(Kur, left_index=True, right_index =True)
B30_EUR
```

### Veriyi özetleme

1.  3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/TRY kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir pivot tablo üretiniz.

```python
B30_EUR["Yıl-Ay"] = B30_EUR.index.strftime("%Y-%m")
B30_EUR
```

```python
B30_EUR.head()
```

```python
pivot = B30_EUR.pivot_table(values =["BIST30", "EUR"], index= "Yıl-Ay", aggfunc="last")
pivot
```

2. Ay kırılımlı olarak ürettiğiniz tabloyu grafik olarak inceleyiniz.


```python
pivot["degisim"] = pivot["BIST30"].diff(1)
pivot["d_yon"] = pivot["degisim"].map(lambda x: "düşen gün"  if x < 0 else "yükselen gün")
pivot["degisim%"] = (pivot["degisim"]/pivot["BIST30"].shift(1)) # % değerlere bakmaknın iyi bir fikir olduğunu videoda aktarmıştık.
pivot
```

```python
px.bar(pivot,y = "degisim%",
       color = "d_yon", color_discrete_map={"yükselen gün":"green","düşen gün":"red"},
       title="BIST30 (1 Ocak 2021 =1)",
       hover_data = {'degisim%' : ':.2f'},
       labels = {'Date' : 'Tarih' },
       text_auto = '.0%')
```

3. Ders sonunda hazırladığınız 2021 ve 2022 yıllarına ait BIST30 ve EUR/USD kur değerlerini içeren bileşke tablodan sadece ayın son işgününe ait değerleri içeren bir tabloyu bir de resample yöntemi ile üretiniz.

```python
Resample= B30_EUR.resample ("BM", convention = 'end').asfreq() #nan çıktı!!!
Resample
```

resample fonksiyonunun yalnızca bazı tipteki dizinlere sahip DataFrame'lerle kullanılabileceği için Veri veri çerçevesinin indeksini bir tarih dizinine dönüştürmemiz gerekiyor. 
Bu kod bu işe yarıyor.

```python
Daily=B30_EUR.resample("d", convention="end").asfreq() #her günü türetmek için
Daily.head(10)
```

4. Resample yöntemi ile ürettiğiniz tabloyu grafik olarak inceleyiniz. 2. adımda elde ettiğiniz grafik ile karşılaştırarak yorumlayınız.

```python
Resample["degisim"] = Resample["BIST30"].diff(1)
Resample["d_yon"] = Resample["degisim"].map(lambda x: "düşen gün"  if x < 0 else "yükselen gün")
Resample["degisim%"] = (Resample["degisim"]/Resample["BIST30"].shift(1)) # % değerlere bakmaknın iyi bir fikir olduğunu videoda aktarmıştık.
Resample
```

```python
px.bar(Resample,y = "degisim%",
       color = "d_yon", color_discrete_map={"yükselen gün":"green","düşen gün":"red"},
       title="BIST30 (1 Ocak 2021 =1)",
       hover_data = {'degisim%' : ':.2f'},
       labels = {'Date' : 'Tarih' },
       text_auto = '.0%')
```
